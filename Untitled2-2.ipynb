```python
from scipy.io import loadmat
train_idxs = loadmat('/Users/mostafahagras/Desktop/PR_data/cuhk03_new_protocol_config_labeled.mat')
import numpy as np
```

```python
import json
with open('/Users/mostafahagras/Desktop/PR_data/feature_data.json', 'r') as f:
    features = json.load(f)
```

```python
labels = train_idxs['labels'] #14096 1_start
camId = train_idxs['camId'] #14096 1_start
filelist = train_idxs['filelist'] #14096
gallery_idx = train_idxs['gallery_idx'] #5328
query_idx = train_idxs['query_idx'] #1400
train_idx = train_idxs['train_idx'] #7368
```

```python
for i in range(len(query_idx)):
    query_idx[i] -=1
```

```python
for i in range(len(gallery_idx)):
    gallery_idx[i] -=1
```

```python
for i in range(len(train_idx)):
    train_idx[i] -=1
```

```python
def return_features(index_id):
    training_features = []
    training_labels = []
    
    for element in gallery_idx:
        if (labels[element[0]] == labels[index_id]):
            if (camId[element[0]] == camId[index_id]):
                continue
        training_features.append(features[element[0]])
        training_labels.append(labels[element[0]][0])
        #print(element[0])
    print(len(training_features))
    #print(training_features)
    return training_features, training_labels

```

```python
from sklearn.neighbors import NearestNeighbors
rank_list = []
for i in query_idx:
    print(i)
    training_features, training_labels = return_features(i)
   # print(i[0])#, training_labels)
    
    #index = i[0] - 1
    neigh = NearestNeighbors(20, metric = 'euclidean')
    neigh.fit(training_features)
    neighbors_return = neigh.kneighbors([features[i[0]]], 20, return_distance = False)
    #for x in neighbors_return:
    #rank_list.append([np.asscalar(labels[index] == labels[training_labels[y]]) for y in neighbors_return])
    x = [np.asscalar(labels[i] == training_labels[y]) for y in neighbors_return.flatten()]
    rank_list.append(x)
```

```python
np.savetxt( 'query_rank_list_final.csv', rank_list, fmt='%i', delimiter= ',' )
```

```python
training_set = []
for i in train_idx:
    training_set.append(features[i[0]])
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA

pca = PCA(n_components = 30)
pca.fit(training_set)
kpca = KernelPCA(n_components = 50, kernel = 'sigmoid')
kpca.fit(training_set)
```

```python
from sklearn.neighbors import NearestNeighbors
pca_rank_list = []
for i in query_idx:
    print(i)
    training_features, training_labels = return_features(i)
   # print(i[0])#, training_labels)
    training_features = kpca.transform(training_features)
    #index = i[0] - 1
    neigh = NearestNeighbors(20, metric = 'euclidean')
    neigh.fit(training_features)
    neighbors_return = neigh.kneighbors(kpca.transform([features[i[0]]]), 20, return_distance = False)
    #for x in neighbors_return:
    #rank_list.append([np.asscalar(labels[index] == labels[training_labels[y]]) for y in neighbors_return])
    x = [np.asscalar(labels[i] == training_labels[y]) for y in neighbors_return.flatten()]
    pca_rank_list.append(x)
```
